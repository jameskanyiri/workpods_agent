```python
from dotenv import load_dotenv

load_dotenv()

from tavily import AsyncTavilyClient

tavily_client = AsyncTavilyClient()
```

```python
query = "How to calculate and present daily and peak load summaries for household appliances and institutions in energy demand analysis?"

response = await tavily_client.search(
    query,
    search_depth="advanced",
    max_results=5,
    include_answer="advanced",
)
print(response)
```
